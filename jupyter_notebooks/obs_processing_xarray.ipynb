### Concat dardar time, big regions

##### JJA

```python
# AFR, WPC
region = "AFR"
```

```python
jja_list = ["Jun", "Jul", "Aug"]
ds_list_jja = [
    xr.open_dataset(file_path + "DARDAR-v3_iwc_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jja_list
]
```

```python
jja_out = file_path + "DARDAR-v3_iwc_JJA2009_{r}.nc".format(r=region)
ds_jja_cat = xr.concat(ds_list_jja, dim="time")
```

```python
ds_jja_cat
```

```python
ds_jja_cat.to_netcdf(jja_out)
```

##### DJF

```python
# AMZ, SPC1, SPC2
region = "SPC2"
```

```python
jf_list = ["Jan", "Feb"]
ds_list_jf = [
    xr.open_dataset(file_path + "DARDAR-v3_iwc_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jf_list
]
```

```python
ds_d = xr.open_dataset(file_path + "DARDAR-v3_iwc_Dec2008_{r}.nc".format(r=region))
```

```python
djf_out = file_path + "DARDAR-v3_iwc_DJF2009_{r}.nc".format(r=region)
ds_djf_cat = xr.concat([ds_d, *ds_list_jf], dim="time")
```

```python
ds_djf_cat
```

```python
ds_djf_cat.to_netcdf(djf_out)
```

### Concat ERA5 time, big regions

##### JJA

```python
# AFR, WPC
region = "AFR"
```

```python
jja_list = ["Jun", "Jul", "Aug"]
temp_ds_list_jja = [
    xr.open_dataset(file_path + "ERA5_T_0.25deg_ml_12-20km_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jja_list
]
zg_ds_list_jja = [
    xr.open_dataset(file_path + "ERA5_zg_0.25deg_ml_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jja_list
]
```

```python
jja_temp_out = file_path + "ERA5_T_0.25deg_ml_12-20km_JJA2009_{r}.nc".format(r=region)
jja_zg_out = file_path + "ERA5_zg_0.25deg_ml_JJA2009_{r}.nc".format(r=region)

temp_ds_jja = xr.concat(temp_ds_list_jja, dim="time")
zg_ds_jja = xr.concat(zg_ds_list_jja, dim="time")
```

```python
temp_ds_jja.to_netcdf(jja_temp_out)
zg_ds_jja.to_netcdf(jja_zg_out)
```

##### DJF

```python
# AMZ, SPC1, SPC2
region = "SPC2"
```

```python
jf_list = ["Jan", "Feb"]
temp_ds_list_jf = [
    xr.open_dataset(file_path + "ERA5_T_0.25deg_ml_12-20km_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jf_list
]
zg_ds_list_jf = [
    xr.open_dataset(file_path + "ERA5_zg_0.25deg_ml_{m}2009_{r}.nc".format(m=month, r=region))
    for month in jf_list
]
```

```python
temp_ds_d = xr.open_dataset(file_path + "ERA5_T_0.25deg_ml_12-20km_Dec2008_{r}.nc".format(m=month, r=region))
zg_ds_d = xr.open_dataset(file_path + "ERA5_zg_0.25deg_ml_Dec2008_{r}.nc".format(m=month, r=region))
```

```python
djf_temp_out = file_path + "ERA5_T_0.25deg_ml_12-20km_DJF2009_{r}.nc".format(r=region)
djf_zg_out = file_path + "ERA5_zg_0.25deg_ml_DJF2009_{r}.nc".format(r=region)

temp_ds_djf = xr.concat([temp_ds_d, *temp_ds_list_jf], dim="time")
zg_ds_djf = xr.concat([zg_ds_d, *zg_ds_list_jf], dim="time")
```

```python
temp_ds_djf.to_netcdf(djf_temp_out)
zg_ds_djf.to_netcdf(djf_zg_out)
```

##### ERA5 T + zg
